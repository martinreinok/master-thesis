```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = 'cathbot-responses.csv'
data = pd.read_csv(file_path)

# Drop the Timestamp and any non-numeric columns
data = data.drop(columns=['Timestamp', 'Any comments for further improvement (optional)'])

# Transform data to long format
data_long = pd.melt(data, var_name='Statement', value_name='Score')

# Set the figure size
# Set the figure size
plt.figure(figsize=(8, 4))

# Create the boxplot with all boxes in the same color
sns.boxplot(data=data_long, x='Statement', y='Score', color='lightblue', whis=[0, 100])

# Rotate x labels for better readability
plt.xticks()

# Set labels and title
plt.xlabel('Statement')
plt.ylabel('Score')
# plt.title('Boxplot of Scores for Each Statement')

# Adjust x-axis to start at 1
plt.gca().set_xticks(range(len(data.columns)))
plt.gca().set_xticklabels(range(1, len(data.columns) + 1))

# Show the plot
plt.tight_layout()
plt.show()
```
